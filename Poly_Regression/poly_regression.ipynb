```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
poly_reg = PolynomialFeatures(degree= 4)
X_poly = poly_reg.fit_transform(X)
line_reg_2 = LinearRegression()
line_reg_2.fit(X_poly, y)
```

# Regular Regression model

```python
plt.scatter(X, y, color = 'orange')
plt.plot(X, lin_reg.predict(X), color = 'purple')
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Fixing the regression model with CEO still attached

```python
plt.scatter(X, y, color = 'orange')
plt.plot(X, line_reg_2.predict(X_poly), color = 'purple')
plt.title("Polynomial Linear Regression (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Smoothing Ridges

```python
X_grid = np.arange(min(X), max(X), .1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'orange')
plt.plot(X_grid, line_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'purple')
plt.title("Polynomial Linear Regression (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
